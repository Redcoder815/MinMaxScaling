```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Titanic-Dataset.csv')
# df.info()
# df.head()

df.duplicated()
cat_col = [col for col in df.columns if df[col].dtype == 'object']
num_col = [col for col in df.columns if df[col].dtype != 'object']

# print('catergory column', cat_col)
# print('numerical column', num_col)

# df[cat_col].nunique()

df1 = df.drop(columns = ['Name','Ticket','Cabin'])
df1.dropna(subset=['Embarked'], inplace=True)
df1['Age'].fillna(df1['Age'].mean(),inplace=True)
round((df1.isnull().sum()/df1.shape[0])*100,2)
# display(df1)
# import matplotlib.pyplot as plt
# plt.boxplot(df1['Age'], vert=False)
# plt.xlabel('Age')
# plt.ylabel('data')
# plt.title('Box plot of age')
# plt.show()

mean = df1['Age'].mean()
std = df1['Age'].std()

lower_bound = mean - 2 * std
upper_bound = mean + 2 * std

df2 = df1[(df1['Age']>=lower_bound) & (df1['Age']<=upper_bound)]
import ipdb
# ipdb.set_trace()
# print('Print after selection')
# print(df2)
X = df2[['Pclass','Sex','Age', 'SibSp','Parch','Fare','Embarked']]
Y = df2['Survived']
# ipdb.set_trace()
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
num_col = [col for col in X.columns if X[col].dtype != object]
ipdb.set_trace()
x1 = X

x1[num_col] = scaler.fit_transform(x1[num_col])
ipdb.set_trace()
x1.head()
```
