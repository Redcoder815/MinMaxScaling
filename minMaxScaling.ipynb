```python
import tensorflow as tf

# Define non-zero values and their indices
values = tf.constant([41, 30], dtype=tf.int32)
indices = tf.constant([[1, 2], [3, 3]], dtype=tf.int64)
dense_shape = tf.constant([4, 8], dtype=tf.int64)

# Create a sparse tensor
sparse_tensor = tf.sparse.SparseTensor(
    indices=indices,
    values=values,
    dense_shape=dense_shape
)

# Print sparse tensor metadata
print("SparseTensor object:")
print(sparse_tensor)

# Convert to dense tensor for visualization
dense_tensor = tf.sparse.to_dense(sparse_tensor)

print("\nDense representation:")
print(dense_tensor.numpy())  # .numpy() for NumPy array output
```
